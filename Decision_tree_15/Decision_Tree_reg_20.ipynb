```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
dataset = pd.read_csv("F:/Study/Wscube-tech/Machine-learning/Datasets/Salary_Data.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
sns.pairplot(data=dataset)
plt.show()
# this graph is linear but we will aplly decision tree on this data 
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Salary"]
```

```python
# train test splotting
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# model 
from sklearn.tree import DecisionTreeRegressor, plot_tree
```

```python
dt = DecisionTreeRegressor(max_depth=5)
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)*100
```

```python
dt.score(x_train,y_train)*100
```

```python
plot_tree(dt)
plt.show()
```

```python
for i in range(1,50 ):

    dt2 = DecisionTreeRegressor(max_depth=i)
    dt2.fit(x_train,y_train)
    print(dt2.score(x_train,y_train),dt2.score(x_test,y_test),i)
    # in case of 2 and 3 there is a less difference 
```
