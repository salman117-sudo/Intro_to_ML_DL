## Frequennt Pattern Growth Algorithm 

```python
import pandas as pd 
import numpy as np 
```

```python
pd.set_option("display.max_rows",500)
```

```python
dataset = pd.read_csv("D:/Study/Wscube-tech/Machine-learning/Datasets/groceries - groceries.csv")
dataset.head(3)
```

```python
dataset.shape
```

```python
market = []
for i in range(0, dataset.shape[0]):
    cus = []
    for j in dataset.columns:
        if type(dataset[j][i]) == str:
            cus.append(dataset[j][i])
    market.append(cus)
## in this code we create a list within a list for removing nan values 
```

```python
l = []
for i in market:
    for j in i:
        l.append(j)
```

```python
import collections
```

```python
p = collections.Counter(l)
```

```python
d = {"Item Name":p.keys(),"values":p.values()}
```

```python
pd.DataFrame(d).sort_values(by=["values"],ascending=False)
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
tr = TransactionEncoder()
tr.fit(market)
```

```python
df=pd.DataFrame(tr.transform(market),columns=tr.columns_)
```

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
fpgrowth(df,min_support=0.05,use_colnames=True,max_len=3).sort_values(["support"])
```
