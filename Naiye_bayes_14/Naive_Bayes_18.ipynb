```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("D:/Study/Wscube-tech/Machine-learning/Datasets/placement11.csv")
dataset.head(3)
```

```python
#sns.kdeplot(data=dataset["iq"])
#plt.show()

# to check normal distribuation 
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="cgpa",y="iq",data=dataset,hue="placement")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100 , gnb.score(x_train,y_train)*100
```

```python
gnb.predict([[5.9,	106	]])
```

```python
plt.figure(figsize=(4,3))

plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

```python
mnb = MultinomialNB()
# this model work good on textual data 
```

```python
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)*100 , mnb.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(4,3))

plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```

```python
bnb = BernoulliNB()
# it show better result in 0 and 1 
```

```python
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)*100  , bnb.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(4,3))

plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```
