## Hierachical clusering (allgomerative, undivisive)

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
dataset = pd.read_csv("F:/Study/Wscube-tech/Machine-learning/Datasets/Iris_raw.csv")
dataset.head(3)
```

```python
#sns.pairplot(data=dataset)
#plt.show()
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,16))
sc.dendrogram(sc.linkage(dataset,method='ward', metric='euclidean'))
plt.savefig("dent.jpg")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=2,linkage="single")
dataset["predict"] = ac.fit_predict(dataset)
```

```python
dataset.head(2)
```

```python
sns.pairplot(data=dataset,hue="predict")
plt.show()
# it will make two cluster  . this cluster work on distance 
```
